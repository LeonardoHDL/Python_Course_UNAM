# Ejercicios de listas y tuplas

## list
Tipo compuesto de dato que puede almacenar distintos elementos (llamados ítems) ordenados entre [ ] y separados con comas.

```python
list1 = [5, 2.1, -9, -520.18]
```

```python
type(list1)
```

```python
# Hacer ejemplos
lista1=[1,2,3,'jeje','hola']
lista2=['hola','como','estas','amigo',1,6,7,3,4]
lista3=[1,2,3,4]
```

## Slicing en listas
Lst[ Initial : End : IndexJump ]

```python
# Hacer ejemplos
print(lista1[1:3])
print(lista1[::-1])
print(lista1[0::2])
```

## Suma de listas
Da como resultado una nueva lista que incluye todos los ítems.

```python
# Hacer ejemplos
print(lista1+lista2)
print(lista1+lista3)
print(lista2+lista3)
```

## Son modificables
A diferencia de las cadenas, en las listas sí podemos modificar sus ítems utilizando índices:

```python
# Hacer ejemplos
```

```python
print(f'lista original:{lista1}')
lista1[2]='siuu'
lista1[4]='niooooo'
print(f'lista modificada:{lista1}')
```

```python
print(f'lista original:{lista2}')
lista2[2]='la maquina'
lista2[4]='arriba el cruzazul'
print(f'lista modificada:{lista2}')
```

## Vaciar lista con [ ]

```python
# Hacer ejemplos
lista=[1,2,3,34,4,5,]
print('lista original:', lista)
lista=[]
print('lista vaciada:', lista)
```

## Función len()

```python
print(lista1)
print('longitud de la lista',len(lista1))
print(lista)
print('longitud de la lista',len(lista))
print(lista2)
print('longitud de la lista',len(lista2))
```

# Listas dentro de listas (anidadas)
Podemos manipular fácilmente este tipo de estructuras utilizando múltiples índices, como si nos refieréramos a las filas y columnas de una tabla.

```python
a = [1,2,3]
b = [4,5,6]
c = [7,8,9]
r = [a,b,c]
```

```python
r[1][1]
```

```python
r[2][2]
```

```python
r[0][2]
```

# Tuple
 Es una colección de elementos distintos ordenados entre ( ) y separados por comas.

```python
tupla1 = (5, 3, 'a')
type(tupla1)
```

Puede haber algunos inconvenientes al tratar de crear una tupla con 1 elemento

```python
tupla2 = (5)
tupla2
```

```python
type(tupla2)
```

```python
tupla2 = (5, )
tupla2
```

```python
type(tupla2)
```

```python
tupla3=(1,5,52,'siuu')
```

### Algunas características de las tuplas son:
- inmutables
- slicing en tuplas
- suma de tuplas
- función len
- multiplicación con enteros

```python
tupla1[1]
```

```python
tupla1[2]='a'
```

```python
tupla1+tupla2
```

```python
len(tupla3)
```

```python
tupla3*5
```

# range function
La función range() retorna una sucesión de numeros en un rango, range(start, stop, step)

```python
x = range(1, 5)
x
```

```python
type(x)
```

```python
x[0]
```

```python
x[-1]
```
