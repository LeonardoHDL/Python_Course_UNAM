1) Calculate the factorial of a number (a non-negative integer)

```python
#1 calculate factorial of non negative number
def calculate_factorial(num):
    num=num
    factorial = 1    
    if num < 0:    
       print(" Factorial does not exist for negative numbers")    
    elif num == 0:    
       print("The factorial of 0 is 1")    
    else:    
       for i in range(1,num + 1):    
           factorial = factorial*i    
    print("The factorial of",num,"is",factorial)

a=int(input('input a non-negative number'))
calculate_factorial(a)
```

3) Given a string, 2 methods to count the spaces in the string

```python
string=input(f'input a sentence')
def check_space(string):
     
    # counter
    count = 0
     
    # loop for search each index
    for i in range(0, len(string)):
         
        # Check each char
        # is blank or not
        if string[i] == " ":
            count += 1
    return count
print(f'number of spaces: {check_space(string)}')
```

```python
string=input(f'input a sentence')
def check_space(string): ## Second method
     
    # counter
    lst=[]
     
    # loop for search each index
    for i in range(0, len(string)):
         
        # Check each value
        # is blank or not
        if string[i] == " ":
            lst.append(0)
    return len(lst)
print(f'number of spaces: {check_space(string)}')
```

4) Given a list, ask for two indexes of the list, interchange those values

```python
def interchanging(lst):
    # Ask for user input for the indexes
    idx1 = int(input("Enter the first index: "))
    idx2 = int(input("Enter the second index: "))
    lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
    # Check if the indexes are valid
    if 0 <= idx1 < len(lst) and 0 <= idx2 < len(lst):
    # Swap values at the given indexes
        print("List after swapping values:", lst)
    else:
        print("Invalid indexes provided. Indexes should be between 0 and", len(lst) - 1)

# Example list
my_list = [10, 20, 30, 40, 50]

interchanging(my_list)
```

6) Given a dictionary, sort the dictionary by key

```python
myDict = {'mate': 10, 'fisica': 9,
        'biologia': 5, 'algebra': 7, 'neuro': 10}
 
Keys = list(myDict.keys())
Keys.sort()
sorted_dict = {i: myDict[i] for i in Keys}
 
print(sorted_dict)
```
